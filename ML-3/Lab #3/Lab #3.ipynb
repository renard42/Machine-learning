Добавляем новые атрибуты - 
1. длина заголовка 
2. количество слов с заглавной буквы (названия различного рода + просто привлекают внимание)
3. количество слов капсом

```python
lemmatizer = WordNetLemmatizer()

titles = news_final['Title'].tolist()
headings = news_final['Headline'].tolist()

def preprosessor(text, headlines=True):
    features = []
    count = 0
    for line in text:
        new_features = []
        toks = word_tokenize(line)
        new_line = [lemmatizer.lemmatize(word) for word in toks if word.isalpha()==True]
        new_features.append(' '.join(new_line))
        new_features.append(len([n for n in toks if n.isdigit()==True]))
        if headlines==False:
            new_features.append(len(toks))
            new_features.append(len([word_2 for word_2 in new_line if word_2.isupper()==True]))
            new_features.append(len([word for word in new_line if word.istitle()==True]))
        else:
            new_features.append(len([word for num, word in enumerate(new_line) if word.istitle()==True and new_line[num-1] not in ['.', '?', '!']]))
            new_features.append(len(sent_tokenize(line)))
        features.append(new_features)
        count+=1
        if headlines==True and count%100==0:
            print('headlines done#', count)
        elif headlines==False and count%100==0:
            print('titles done#', count)
    return features

feats_titles = preprosessor(titles, headlines=False)
feats_headings = preprosessor(headings)
```

```python
titles = pd.DataFrame(data=feats_titles, columns=['Title', 'Title_nums', 'Title_len', 'Title_Caps', 'Title_Titles'])
headings = pd.DataFrame(data=feats_headings, columns=['Heads', 'Head_nums', 'Head_Titles', 'Head_Sents'])
```
